# Inkscape es más que suficient para Ingeniería en papel

**Inkscape es más que suficiente porque actúa como el "núcleo central" del proyecto:** te permite diseñar la ingeniería de papel (**pliegues en V y paralelos**), crear el **diseño gráfico impreso** (**el marcador de RA**) e **integrar directamente los códigos QR vectoriales** que conectarán el papel físico con los mundos virtuales.<br>

## Intergación Inkscape
La integración de **Inkscape** con tecnologías web inmersivas de código abierto (**WebXR, A-Frame y Three.js**) es una de las estrategias más potentes para conectar el **diseño editorial físico** con el **desarrollo web en 3D**.En este ecosistema, **Inkscape** deja de ser un simple programa de dibujo y se transforma en un **generador de assets vectoriales interactivos**, **mapeado de texturas (UV)** y **generador de interfaces de usuario (UI)** que las librerías de JavaScript pueden renderizar directamente en el navegador.A continuación, se detallan las referencias de uso, los flujos de integración técnicos y el comportamiento del software en este flujo:

### 1. Inkscape como Editor de Texturas y Mapas UV para Three.js

En el desarrollo con **Three.js y WebXR**, los objetos **3D necesitan texturas para verse reales**. Los desarrolladores utilizan **Inkscape** para pintar el "despliegue" (**UV Map**) de los planos del libro pop-up. 
- **Flujo de Texturizado:** Los alumnos crean las ilustraciones en Inkscape respetando las dimensiones exactas de los pliegues en V o paralelos. Esta imagen se exporta como .PNG o .JPG de alta resolución.
  
- **Implementación en Código:** Dentro del script de Three.js, esa ilustración vectorial hecha en **Inkscape** se carga como un mapa de textura para aplicarla a un plano 3D (THREE.MeshBasicMaterial o **THREE.MeshStandardMaterial**), logrando que la versión digital que se ve en las gafas WebXR o en el celular sea idéntica al papel impreso.

```javascript
// Ejemplo en Three.js cargando el arte hecho en Inkscape
const textureLoader = new THREE.TextureLoader();
const inkscapeTexture = textureLoader.load('textura_popup_inkscape.png');
const material = new THREE.MeshBasicMaterial({ map: inkscapeTexture });

```

### 2. Inkscape para el Diseño de Interfaces 2D (UI) en Entornos de Realidad Virtual (WebXR)

Uno de los mayores retos en **WebXR y A-Frame** es crear menús, botones o textos flotantes legibles dentro del espacio tridimensional. Las fuentes de texto nativas en 3D consumen muchos recursos del procesador.

- **La Solución con Inkscape:** Los alumnos diseñan los menús interactivos, la iconografía y los textos explicativos en **Inkscape**.
- Integración en A-Frame: **A-Frame** permite renderizar imágenes planas directamente en el espacio 3D usando la etiqueta <a-image> o utilizándolas como texturas de fondo para componentes interactivos de UI. Esto garantiza que el texto se vea perfectamente nítido en pantallas móviles y visores VR sin importar la resolución.


### 3. El Súper Poder Oculto: Cargar Archivos .SVG de Inkscape directamente en el código 3D

Dado que Inkscape guarda sus archivos en formato SVG nativo (un estándar web basado en código XML), tanto **Three.js como A-Frame pueden leer el archivo de Inkscape directamente desde el código sin necesidad de convertirlo en imagen estática**

- **SVGLoader de Three.js**: Three.js cuenta con una herramienta oficial llamada **SVGLoader**. Esta herramienta lee las líneas y curvas vectoriales diseñadas en **Inkscape** y las **transforma automáticamente en geometrías 3D (THREE.ShapeGeometry) en tiempo real**.

- **Animación por Código**: Al cargar el archivo de Inkscape directamente como vectores en la web, los alumnos pueden **programar mediante código que los trazados cambien de color, se deformen o brillen cuando el usuario interactúe con ellos** en Realidad Virtual, algo imposible si se usara una imagen fija.

## Referencias en internet de Inkscape
### 1. El manual académico de Ingeniería de Papel con Inkscape

Una de las referencias teóricas y prácticas más completas disponibles de forma libre es el libro digital del Profesor Klein titulado Paper Engineering & Pop-ups (PDF).

http://www.arts.professorklein.com/docs/pop_up_books/Paper%20Engineering.pdf

- **De qué trata:** Es una guía académica diseñada específicamente para enseñar a las personas a estructurar sus propios proyectos de **ingeniería de papel**.

- **Uso de Inkscape:** Dedica capítulos enteros (como el capítulo 14) a Inkscape como el software principal de dibujo. Explica paso a paso cómo configurar el programa desde cero y cómo utilizar las herramientas vectoriales de precisión para trazar los planos de corte que darán volumen a los libros.

### 2. Videotutoriales prácticos de Mecanismos Pop-Up en Inkscape

Para la parte técnica y de manufactura (impresión y corte automatizado), la comunidad de artesanos digitales cuenta con guías detalladas:

- **Creación de archivos SVG Pop-Up desde cero:** El videotutorial How to Make a **DIY Pop Up Card SVG File for Free in Inkscape** es una referencia indispensable de la creadora Dinosaur Mama. Aunque se enfoca en tarjetas, la lógica geométrica es exactamente la misma que la de un libro.

  https://www.youtube.com/watch?v=lWNmOg8XdIE

  - Lo que enseña: Muestra cómo usar los nodos de **Inkscape** para crear **líneas de perforación uniformes y guías de plegado (estilo parallel folds y escalones)**. Explica cómo duplicar trazos y usar la función de Combinar (Combine) en el menú de trayectos para estructurar las pestañas sin que se distorsionen los vectores al cerrar el mecanismo.

- **Guías escritas y plantillas:** La misma autora complementa el video con su artículo técnico How to Make a Pop Up Card SVG for Free in Inkscape, donde incluye capturas de pantalla de la interfaz de Inkscape para configurar el grosor de los trazos, vectorizar imágenes y alinear los elementos perfectamente al eje o "valle" central donde dobla el libro.

# Blender para construir Libros Pop-up -  Ingeniería en papel

Las principales referencias de internet para diseñar libros pop-up en Blender provienen de creadores independientes que combinan herramientas de física rígida (**Rigid Body**), esqueletos (**Armatures**) y el lápiz de cera (**Grease Pencil**).

## 1. El canal de Sophie Jantak (Pionera en el método interactivo

Es la referencia moderna más citada sobre el tema. Desarrolló un flujo de trabajo para crear pop-ups físicos reales usando las herramientas de simulación e ilustración 2D/3D de Blender <br>

- **Video Concepto**: I want to make pop-up books! (with Blender). En este video explica cómo pasó de querer ver su arte digital impreso a usar las físicas rígidas de Blender para probar si los pliegues cierran correctamente sin colisionar.

  https://www.youtube.com/watch?v=WoKEZeCj9QA

- **Video Técnico:** The BASICS of creating realistic pop ups in Blender. Muestra el paso a paso de cómo configurar las uniones del papel, dibujar sobre los planos en 3D usando Grease Pencil

  https://www.youtube.com/watch?v=0nN4EKwnx8M

- **Ejemplo en formato corto:** Starting to design a pop up card in Blender! es un resumen rápido de cómo calcula pliegues paralelos simulando la gravedad y los dobleces físicos

https://www.youtube.com/shorts/vWu7uxxOeSE

## 2. El método clásico con esqueletos (Armatures/Rigging)

Si en lugar de simulaciones físicas prefieres controlar la apertura del libro de forma matemática exacta con palancas digitales, la comunidad recurre al sistema de huesos (rigging):

- **Tutorial de Sardi Pax:** Documentado en plataformas como Lesterbanks, es un clásico que enseña a crear y animar un libro pop-up utilizando esqueletos (Armatures) de Blender. Al rotar un "hueso" maestro, el mecanismo despliega las páginas y las estructuras de forma fluida.

https://lesterbanks.com/2014/02/rigging-animating-popup-book-blender/

- **Hilos de comunidad:** Foros de discusión técnica como Blender Stack Exchange: Rigging for Popup Book explican con diagramas cómo configurar restricciones cinemáticas inversas (IK constraints) para obligar a las pestañas a quedarse "pegadas" a las páginas mientras el libro se abre en 180 o 90 grados.

https://blender.stackexchange.com/questions/207766/rigging-for-popup-book


## 3. La extensión oficial para "Aplanar" el papel (Export Paper Model)

Una vez que el mecanismo funciona en la pantalla, necesitas convertir ese objeto tridimensional en un plano 2D para imprimir y cortar. La herramienta estándar es una extensión oficial de Blender: <br>

- **Extensión:** El add-on Export Paper Model en Blender Extensions (desarrollado originalmente por Adam Dominec).

https://extensions.blender.org/add-ons/export-paper-model/

- **Documentación oficial:** El Manual de Blender: Paper Model detalla cómo usar el menú Mesh ‣ Unfold para marcar costuras en las aristas del papel y exportar las plantillas directamente en formatos vectoriales limpios como .pdf o .svg.

https://docs.blender.org/manual/es/2.91/addons/import_export/paper_model.html

- **Tutorial de soporte:** Canales enfocados en maquetación como Stahlhart Papercraft ofrecen guías específicas como Blender basics for paper modelers – export and seams

https://www.youtube.com/watch?v=2qwZGWjhtcM
